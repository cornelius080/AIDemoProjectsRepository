# EXTRACT AGENT

```python
from typing import Literal
from langchain.tools import tool
from langchain_tavily import TavilyExtract

@tool
def tavily_extract_no_structured(
    url: str,
    extract_depth: Literal["basic", "advanced"] = "basic",
    chunks_per_source: Literal[1, 2, 3, 4, 5] = 3,
    include_images: bool = False,
    include_favicon: bool = False,
    format_text: Literal["markdown", "text"] = "markdown"
) -> str:
    """
    Extract content from a web page using TavilyExtract.

    Args:
        url: The URL of the page to extract content from.
        extract_depth: Extraction depth ("basic" or "advanced").
        chunks_per_source: Number of chunks per source (1-5).
        include_images: Whether to include images (True or False).
        include_favicon: Whether to include favicon (True or False).
        format_text: Output format ("markdown" or "text").

    Returns:
        Extracted content as a string.
    """

    tool_instance = TavilyExtract(
        extract_depth=extract_depth,
        chunks_per_source=chunks_per_source,
        include_images=include_images,
        include_favicon=include_favicon,
        format=format_text,
    )

    return tool_instance.invoke({"urls": [url]})
```

```python
from langchain.agents import create_agent
from langchain_ollama import ChatOllama

model = ChatOllama(model="qwen3.5:9b")

agent = create_agent(
    model=model,
    tools=[tavily_extract_no_structured],
    #tools=[tavily_extract_single_page],
    #tools=[tavily_extract],
    system_prompt = """You are a helpful assistant. 
                        Retrieve the entire original content from the user defined webpage. 
                        Do not modify it.
                        Make sure to use the 'tavily_extract' tool with the following parameters:
                        - extract_depth = "basic"
                        - chunks_per_source = 3
                        - include_images = False
                        - include_favicon = False
                        - format_text = "markdown" """
)
```

```python
from langchain.messages import HumanMessage

query = HumanMessage(content=[
    {"type": "text", "text": "https://www.soniaperonaci.it/pasta-alla-carbonara/"},
])

result = agent.invoke({"messages": [query]})
```

# EXTRACT AGENT - STRUCTURED OUTPUT: SINGLE PAGE APPROACH

```python
from pydantic import BaseModel, HttpUrl, Field
from typing import List

class TavilyExtractOutput(BaseModel):
    """Structured output for a single webpage extraction."""
    url: HttpUrl = Field(description="The URL of the webpage")
    title: str = Field(description="The title of the webpage")
    raw_content: str = Field(description="The extracted raw content")
    images: List[str] = Field(default_factory=list, description="List of image URLs found on the page")
```

```python
from langchain.tools import tool
from typing import Literal
from langchain_tavily import TavilyExtract

@tool
def tavily_extract_single_page(
    url: str,
    extract_depth: Literal["basic", "advanced"] = "basic",
    chunks_per_source: Literal[1, 2, 3, 4, 5] = 3,
    include_images: bool = False,
    include_favicon: bool = False,
    format_text: Literal["markdown", "text"] = "markdown"
) -> TavilyExtractOutput:
    """
    Extract content from a web page using TavilyExtract.

    Args:
        url: The URL of the page to extract content from.
        extract_depth: Extraction depth ("basic" or "advanced").
        chunks_per_source: Number of chunks per source (1-5).
        include_images: Whether to include images (True or False).
        include_favicon: Whether to include favicon (True or False).
        format_text: Output format ("markdown" or "text").
        
    Returns:
        Structured output with Pydantic model TavilyExtractOutput.
    """
    # Initialize the TavilyExtract instance
    extractor = TavilyExtract(
        extract_depth=extract_depth,
        chunks_per_source=chunks_per_source,
        include_images=include_images,
        include_favicon=include_favicon,
        format=format_text,
    )

    # Invoke the extraction
    raw_result = extractor.invoke({"urls": [url]})

    # Extract the first result (single-page approach)
    item = raw_result.get("results", [])[0]

    # Return a structured output
    return TavilyExtractOutput(
        url=item["url"],
        title=item["title"],
        raw_content=item["raw_content"],
        images=item.get("images", [])
    )
```

```python
from langchain.agents import create_agent
from langchain_ollama import ChatOllama

model = ChatOllama(model="qwen3.5:9b")

agent = create_agent(
    model=model,
    #tools=[tavily_extract_no_structured],
    tools=[tavily_extract_single_page],
    #tools=[tavily_extract],
    system_prompt = """You are a helpful assistant. 
                        Retrieve the entire original content from the user defined webpage. 
                        Do not modify it.
                        Make sure to use the 'tavily_extract' tool with the following parameters:
                        - extract_depth = "basic"
                        - chunks_per_source = 3
                        - include_images = False
                        - include_favicon = False
                        - format_text = "markdown" """
)
```

```python
from langchain.messages import HumanMessage

query = HumanMessage(content=[
    {"type": "text", "text": "https://www.soniaperonaci.it/pasta-alla-carbonara/"},
])

result = agent.invoke({"messages": [query]})
```

# EXTRACT AGENT: STRUCTURED OUTPUT - MULTI PAGE APPROACH

```python
from pydantic import BaseModel, HttpUrl, Field
from typing import List

class TavilyExtractOutputItem(BaseModel):
    """Structured output for a single webpage extraction."""
    url: HttpUrl = Field(description="The URL of the webpage")
    title: str = Field(description="The title of the webpage")
    raw_content: str = Field(description="The extracted raw content")
    images: List[str] = Field(default_factory=list, description="List of image URLs found on the page")

class TavilyExtractOutput(BaseModel):
    """Represents the structured output containing all results."""
    results: List[TavilyExtractOutputItem] = Field(description="List of extracted results")
```

```python
from typing import Literal
from langchain.tools import tool
from langchain_tavily import TavilyExtract

@tool
def tavily_extract(
    url: str,
    extract_depth: Literal["basic", "advanced"] = "basic",
    chunks_per_source: Literal[1, 2, 3, 4, 5] = 3,
    include_images: bool = False,
    include_favicon: bool = False,
    format_text: Literal["markdown", "text"] = "markdown"
) -> TavilyExtractOutput:
    """
    Extract content from a web page using TavilyExtract.
    Returns a structured output using TavilyExtractOutput Pydantic model.
    
    Args:
        url: The URL of the page to extract content from.
        extract_depth: Extraction depth ("basic" or "advanced").
        chunks_per_source: Number of chunks per source (1-5).
        include_images: Whether to include images (True or False).
        include_favicon: Whether to include favicon (True or False).
        format_text: Output format ("markdown" or "text").
        
    Returns:
        TavilyExtractOutput: Structured output with list of results.
    """
    # Initialize the TavilyExtract instance
    tool_instance = TavilyExtract(
        extract_depth=extract_depth,
        chunks_per_source=chunks_per_source,
        include_images=include_images,
        include_favicon=include_favicon,
        format=format_text,
    )

    # Invoke the extraction
    raw_result = tool_instance.invoke({"urls": [url]})

    # Parse results into structured Pydantic models
    structured_results = [
        TavilyExtractOutputItem(
            url=item["url"],
            title=item["title"],
            raw_content=item["raw_content"],
            images=item.get("images", [])
        )
        for item in raw_result.get("results", [])
    ]

    # Return the wrapped output
    return TavilyExtractOutput(results=structured_results)
```

```python
from langchain.agents import create_agent
from langchain_ollama import ChatOllama

model = ChatOllama(model="qwen3.5:9b")

agent = create_agent(
    model=model,
    #tools=[tavily_extract_no_structured],
    #tools=[tavily_extract_single_page],
    tools=[tavily_extract],
    system_prompt = """You are a helpful assistant. 
                        Retrieve the entire original content from the user defined webpage. 
                        Do not modify it.
                        Make sure to use the 'tavily_extract' tool with the following parameters:
                        - extract_depth = "basic"
                        - chunks_per_source = 3
                        - include_images = False
                        - include_favicon = False
                        - format_text = "markdown" """
)
```

```python
from langchain.messages import HumanMessage

query = HumanMessage(content=[
    {"type": "text", "text": "https://www.soniaperonaci.it/pasta-alla-carbonara/"},
])

result = agent.invoke({"messages": [query]})
```
