You can instantiate a ChatHuggingFace model in two different ways, either from a HuggingFaceEndpoint or from a HuggingFacePipeline.

# HuggingFaceEndpoint

```python
from langchain_huggingface import ChatHuggingFace, HuggingFaceEndpoint
import os, getpass

custom_env_var = "HF_TOKEN_READ"

if not os.environ.get(custom_env_var):
  os.environ[custom_env_var] = getpass.getpass("Enter API key for HuggingFace: ")

llm = HuggingFaceEndpoint(
    repo_id="MiniMaxAI/MiniMax-M2",
    huggingfacehub_api_token=os.environ.get(custom_env_var),
    temperature=0.7,
    top_p=0.95,
    max_new_tokens=2000,
    # task="text-generation",
    # max_new_tokens=512,
    # do_sample=False,
    # repetition_penalty=1.03,
    # provider="auto",  # let Hugging Face choose the best provider for you
)

model_endpoint = ChatHuggingFace(llm=llm)
```

### DEFINE THE AGENT

```python
from langchain.agents import create_agent
from langgraph.checkpoint.memory import InMemorySaver 

agent_endpoint = create_agent(
    model = model_endpoint,
    checkpointer=InMemorySaver(),  
)
```

```python
question = [
    {"role": "system", "content": "You are a helpful assistant."},
    {"role": "user", "content": "Hello my name is Mark and my favourite colour is green."},
]
config = {"configurable": {"thread_id": "1"}}
```

```python
for token, metadata in agent_endpoint.stream(  
    {"messages": question},
    config,
    stream_mode="messages",
):
    #print(f"node: {metadata['langgraph_node']}")
    print(token.content, end="", flush=True)
```

# HuggingFacePipeline

```python
from langchain_huggingface import ChatHuggingFace, HuggingFacePipeline

llm = HuggingFacePipeline.from_model_id(
    model_id="HuggingFaceTB/SmolLM2-360M-Instruct",
    task="text-generation",
    pipeline_kwargs=dict(
        max_new_tokens=512,
        # do_sample=False,
        # repetition_penalty=1.03,
    ),
)

model_pipeline = ChatHuggingFace(llm=llm)
```

### DEFINE THE AGENT

```python
from langchain.agents import create_agent
from langgraph.checkpoint.memory import InMemorySaver 

agent_pipeline = create_agent(
    model = model_pipeline,
    checkpointer=InMemorySaver(),  
)
```

```python
question = [
    {"role": "system", "content": "You are a helpful assistant."},
    {"role": "user", "content": "Hello my name is Mark and my favourite colour is green. Are you ready for some chatting?"},
]
config = {"configurable": {"thread_id": "1"}}
```

```python
for token, metadata in agent_pipeline.stream(  
    {"messages": question},
    config,
    stream_mode="messages",
):
    #print(f"node: {metadata['langgraph_node']}")
    print(token.content, end="", flush=True)
```
